아래 데이터를 참조하여 18/19년도에 대해서 Price to Dream Ratio (PDR)을 계산하세요.

변수 이름은 df2, feature 이름은 PDR로, df2['PDR']을 통해 ~ 값이 확인 되어야합니다.

```python
#PDR = 기업가치(시가총액) / TAM(전체시장규모) * 시장점유율
#시가총액 = PER / 순이익
#PER = 시가총액 / 순이익
#전제: 일반담배기준 점유율 18년 : 62%, 19년 : 63.5% / 시장규모 18년 : 15조, 19년 15조


col = ['구분', '종가', '발행주식수', '시가총액', '시장점유율', 'PDR']
ind = [0, 1]
con = [['18년', 101500, 137292497,13935188445500,62.0, ''], ['19년','93800','137292497','12878036218600','63.5','']]
df2 = pd.DataFrame(con, columns=col, index=ind)
df2
```

```python
df2.info()
```

```python
df2["종가"] = df2["종가"].apply(pd.to_numeric, errors = 'coerce')
df2["발행주식수"] = df2["발행주식수"].apply(pd.to_numeric, errors = 'coerce')
df2["시가총액"] = df2["시가총액"].apply(pd.to_numeric, errors = 'coerce')
df2["시장점유율"] = df2["시장점유율"].apply(pd.to_numeric, errors = 'coerce')
df2["PDR"] = df2["PDR"].apply(pd.to_numeric, errors = 'coerce')

df2.dtypes
```

```python
df2['PDR'] = (df2['시가총액']/ 15000000000000) * 0.62
```
